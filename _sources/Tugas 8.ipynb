```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.datasets import make_regression
from matplotlib import pyplot
X_test, y_test = make_regression(n_samples=20, n_features=1, noise=10)
reg = LinearRegression().fit(X_test, y_test)
reg.score(X_test, y_test)
# reg.coef_
# reg.intercept_
# reg.predict(np.array([[3]]))
```

```python
df = pd.DataFrame(X_test, y_test)
df
```

```python
x = X_test.tolist()
y = y_test.tolist()
```

```python
#nilai koefisien (a)
reg.coef_
```

```python
#intercept baris (b)
reg.intercept_
```

```python
# mencari y dengan memasukan x=3 
reg.predict(np.array([[3]]))
```

```python
pyplot.scatter(X_test,y_test)
Y_plot = reg.coef_*X_test+reg.intercept_

pyplot.plot(X_test, Y_plot, color='r')
pyplot.show()
```
