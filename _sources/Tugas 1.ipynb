# **Membaca data dari URL**

```python
import pandas as pd

url='https://vincentarelbundock.github.io/Rdatasets/csv/AER/HMDA.csv'
c=pd.read_csv(url)
print(c)
```

```python
c.info()
```

**Data Rasio Pembayaran terhadap Pendapatan.**

```python
data_pirat = c['pirat']
data_pirat
```

**Data Biaya Perumahan untuk Rasio Pendapatan.**

```python
data_hirat =c['hirat']
data_hirat
```

**Data Rasio Pinjaman terhadap Nilai.**

```python
data_lvrat =c['lvrat']
data_lvrat
```

# **Membuat statistik diskriptif dari data (setiap atribut yang bertipe numerik)**

```python
import statistics as st
```

**Median, Modus, Mean dari Data Rasio Pembayaran terhadap Pendapatan.**

Median

```python
st.median(data_pirat)
```

Modus

```python
st.mode(data_pirat)
```

Mean

```python
st.mean(data_pirat)
```

**Median, Modus, Mean dari Data Biaya Perumahan untuk Rasio Pendapatan.**

Median

```python
st.median(data_hirat)
```

Modus

```python
st.mode(data_hirat)
```

Mean

```python
st.mean(data_hirat)
```

**Median, Modus, Mean dari Data Rasio Pinjaman terhadap Nilai.**

Median

```python
st.median(data_lvrat)
```

Modus

```python
st.mode(data_lvrat)
```

Mean

```python
st.mean(data_lvrat)
```

# **Mendiskripsikan kemiringan/skewned dari data setiap variabel (bertipe numerik)**

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

**Skewned Data Rasio Pembayaran terhadap Pendapatan.**

```python
data_pirat.skew()
```

```python
plt.figure()
sns.distplot(data_pirat)
plt.show()
```

**Skewned Data Biaya Perumahan untuk Rasio Pendapatan.**

```python
data_hirat.skew()
```

```python
plt.figure()
sns.distplot(data_hirat)
plt.show()
```

**Skewned Data Rasio Pinjaman terhadap Nilai.**

```python
data_lvrat.skew()
```

```python
plt.figure()
sns.distplot(data_lvrat)
plt.show()
```
